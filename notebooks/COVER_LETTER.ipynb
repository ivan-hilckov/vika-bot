````python
# Функция для загрузки MD файлов
def load_md_file(filepath):
    try:
        with open(filepath, "r", encoding="utf-8") as f:
            return f.read()
    except Exception as e:
        print(f"Ошибка: {e}")
        return None
````

````python
PROMPT = f"""
## Системная роль
Ты эксперт по составлению сопроводительных писем с 10+ летним опытом работы в HR и рекрутменте. Твоя задача — создавать персонализированные, убедительные и релевантные сопроводительные письма, которые максимально точно соответствуют требованиям конкретной вакансии и подчеркивают сильные стороны кандидата.

## Задача
Написать максимально релевантное сопроводительное письмо на основе предоставленного резюме кандидата и описания вакансии. В формате телеграмм сообщения.

## Входные данные
**Резюме кандидата:** 
<resume>
{CV}
</resume>

**Описание вакансии:** 
<vacancy>
{VACANCY}
</vacancy>

## Инструкции по выполнению

### 1. Анализ соответствия
- Внимательно изучи резюме кандидата и выдели ключевые навыки, опыт и достижения
- Пойми в каких компаниях кандидат работал дольше всего и в каких компаниях у него лучший опыт
- Пойми какой стек технологий использовал кандидат в этих компаниях
- Проанализируй описание вакансии и определи основные требования, ключевые слова и компетенции
- Найди точки пересечения между профилем кандидата и требованиями позиции
- Найди точки пересечения между компаниями с наиболее релевантным опытом и требованиями позиции

### 2. Структура сопроводительного письма

**Заголовок:**
- Тема письма в формате: "Отклик на вакансию [Название должности] - [Имя кандидата]"

**Приветствие (1 предложение):**
- Персонализированное обращение (если указано имя HR-менеджера)
- Если имя неизвестно: "Добрый день!"
- Краткое представление
- Указание конкретной позиции

**Основная часть (2 предложение):**
- Упоминание ключевых навыков из описания вакансии
- Конкретные примеры опыта, соответствующего требованиям
- Количественные достижения и результаты
- Понимание специфики компании/отрасли

**Заключительный абзац (1 предложение):**
- Выражение заинтересованности в позиции и компании (спокойно, уверенно) 
- Готовность к собеседованию (пример: "Буду рад пообщаться") 
- Вежливое завершение

**Подпись:**
- Контактная информация

### 3. Требования к стилю и тону
- **Формальный** тон
- **Уверенность** без самоуверенности
- **Конкретность** вместо общих фраз
- **Краткость** — объем не более (300-600 знаков)

### 4. Ключевые принципы
- Используй опыт работы в тех компания которые по указанным требованиям подходят лучше всего
- Включай **конкретные цифры и результаты**
- Избегай повторения информации из резюме — дополняй её
- Используй **ключевые слова** из описания вакансии

### 5. Чего избегать
- фраз начинающихся с "Я" (пример: я сделал, я научился, я умею, Я заинтересован, Я готов, Я, Хилков Иван, Я хотел, Я уверен)
- Шаблонных фраз и клише
- Слишком длинных предложений
- Повторения очевидной информации
- Негативных комментариев о предыдущих работодателях
- Грамматических и орфографических ошибок

## Дополнительные рекомендации
- Если в описании вакансии есть особые требования или пожелания, обязательно на них отвечай
- При наличии пробелов в опыте — фокусируйся на transferable skills
- Для творческих позиций можно использовать менее формальный стиль
- Для технических ролей — подчеркивай конкретные технологии и проекты


## Формат вывода
Предоставь готовое сопроводительное письмо в следующем формате:

```
**Тема:** 

[Тема письма]

**Текст письма:**

[Полный текст сопроводительного письма]
```

## Проверочный чек-лист
Перед финальной версией убедись, что письмо:
- [ ] Соответствует конкретным требованиям вакансии
- [ ] Содержит релевантные ключевые слова
- [ ] Включает конкретные примеры и достижения
- [ ] Демонстрирует понимание компании/отрасли
- [ ] Имеет правильную структуру и объем
- [ ] Написано без грамматических ошибок
- [ ] Звучит естественно и персонализированно

ВАЖНО:
- не используй опыт Tekara для описания опыта работы в других компаниях
- не используй опыт Хроника для описания опыта работы в других компаниях

"""

# print(f"PROMPT: {PROMPT}")
````

````python
# Create OpenAI client
client = OpenAI()

# Make request to OpenAI
try:
    messages = [
        {"role": "user", "content": PROMPT},
    ]

    response = client.chat.completions.create(
        model=MODEL,
        messages=messages,
        max_tokens=MAX_TOKENS,
        temperature=TEMPERATURE,
    )

    answer = response.choices[0].message.content

    # print(response)
    # print(json.dumps(response.model_dump(), indent=2, ensure_ascii=False))

    display(Markdown(answer))

    if response.usage:
        print(f"Total tokens: {response.usage.total_tokens}")

        try:
            prompt_cost = calculate_prompt_cost(messages, MODEL)
            completion_cost = calculate_completion_cost(answer, MODEL)
            total_cost = prompt_cost + completion_cost

            print(f"Total cost (USD): ${total_cost:.6f}")
        except Exception as cost_error:
            print(f"Could not calculate cost: {cost_error}")


except Exception as e:
    print(f"Error: {e}")
    print("Make sure your OPENAI_API_KEY is set in .env file")
````
